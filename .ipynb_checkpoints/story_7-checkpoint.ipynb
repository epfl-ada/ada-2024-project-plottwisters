```python
import pandas as pd 
import numpy as np 
import re 
import seaborn as sns
import matplotlib.pyplot as plt 
from bs4 import BeautifulSoup 
from datetime import datetime
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import plotly.express as px
import ast
```

```python
folder = "generated/"

df_movies=pd.read_csv(folder + "USE_THIS_DATASET.csv", index_col = 'Wikipedia movie ID')
df_movies['decade'] = (df_movies['Movie release year'] // 10) * 10

def month_to_season(month):
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Spring"
    elif month in [6, 7, 8]:
        return "Summer"
    elif month in [9, 10, 11]:
        return "Fall"

# Apply the function to create a 'season' column
df_movies['season'] = df_movies['Movie release month'].apply(month_to_season)


df_movies.sample(5)
```

```python
df_2 = df_movies[df_movies['Movie countries'].apply(lambda x: 'United States of America' in x)]
df_2['Movie genres'] = df_2['Movie genres'].apply(ast.literal_eval)
df_exploded = df_2.explode('Movie genres')
#df_exploded.sample(5)
```

```python
top_10_genres = df_exploded['Movie genres'].value_counts().head(10).index.tolist()
top_10_genres

filtered_movies = df_exploded[df_exploded['Movie genres'].isin(top_10_genres)]
filtered_movies.sample(5)
```

## Seasonality Score

Calculate the total box office revenue for each genre, grouped by release season:

```python
# Group by genre and season, summing revenue
df_filtered = filtered_movies.dropna(subset=['Movie genres', 'Movie box office revenue','season'])
df_filtered['Movie genres'] = df_filtered['Movie genres'].str.split(',').str[0]

genre_seasonal_revenue = df_filtered.groupby(['Movie genres', 'season'])['Movie box office revenue'].sum().reset_index()
genre_seasonal_revenue.sample(5)
```

For each genre, calculate the percentage of its annual box office revenue that occurs in each seasons:

```python
# Total yearly revenue per genre
total_genre_revenue = genre_seasonal_revenue.groupby('Movie genres')['Movie box office revenue'].transform('sum')

# Calculate the percentage share of revenue per month for each genre
genre_seasonal_revenue['seasonal_share'] = genre_seasonal_revenue['Movie box office revenue'] / total_genre_revenue
genre_seasonal_revenue.sample(5)
```

```python
# Calculate variance of seasonal shares for each genre

seasonality_scores = genre_seasonal_revenue.groupby('Movie genres')['seasonal_share'].var().reset_index()
seasonality_scores.columns = ['Movie genres', 'seasonality_score']
seasonality_scores.sample(5)



```

```python
seasonality_scores.sort_values('seasonality_score', ascending=False).plot(
    x='Movie genres', y='seasonality_score', kind='bar', title="Seasonality Score by Genre"
)
```

High Seasonality Score (High Variance or Low Entropy): The genres success is concentrated in specific months (e.g., Christmas movies in December).
Low Seasonality Score (Low Variance or High Entropy): The genres success is evenly distributed throughout the year.

```python
top_genres = genre_seasonal_revenue['Movie genres'].unique()  # Top 5 genres

for genre in top_genres:
    data = genre_seasonal_revenue[genre_seasonal_revenue['Movie genres'] == genre]
    plt.plot(data['season'], data['seasonal_share'], label=genre)

plt.legend()
plt.title("seasonal Revenue Distribution for Top Genres")
plt.xlabel("Season")
plt.ylabel("seasonal Share")
plt.show()
```
