```python
import pandas as pd
import numpy as np
import pandas as pd
import json
import time
import numpy as np
import understatapi
from sqlalchemy import create_engine
import re
import requests
from bs4 import BeautifulSoup 
from pydantic import BaseModel
from typing import List, Optional
from selenium import webdriver #open chrome
from supabase import create_client, Client
import codecs
import psycopg2
from datetime import datetime
from typing import Literal
```

```python
with open('data_events.txt', encoding='utf-8') as file:
    html_code = file.read()
soup = BeautifulSoup(html_code, 'html.parser')


dates = []
titles = []
descriptions = []

entries = soup.find_all('li', class_='list-timeline__item')

for entry in entries:
    date_div = entry.find('div', class_='timeline-item__date')
    date = date_div.time.get_text(strip=True) if date_div and date_div.time else "No date found"
    dates.append(date)
    
    title_tag = entry.find('h1', class_='timeline-item__title')
    title = title_tag.get_text(strip=True) if title_tag else "No title found"
    titles.append(title)
    
    body_div = entry.find('div', class_='timeline-item__body')
    body = body_div.get_text(strip=True) if body_div else "No description found"
    descriptions.append(body)

print("Dates:", dates)
print("Titles:", titles)
print("Descriptions:", descriptions)
```

```python
df = pd.DataFrame({
        'Date': dates,
        'Title': titles,
        'Description': descriptions
    })

df['Date'] = pd.to_datetime(df['Date'], format="%b %d, %Y")
```
